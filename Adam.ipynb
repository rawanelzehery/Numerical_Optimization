```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X = np.linspace(0,20)
X
```

```python
a=-1
b=2
y=a*X+b
y
```

```python
plt.scatter(X,y)
```

```python
def batch(x,y,a,mx):
    th0 = 0
    th1 = 0
    m = y.shape[0]
    cost_list = []
    th0l = []
    th1l = []
    y_p = []
    for i in range(mx):
        
        print("****************",i,"**************")
        th0l.append(th0)
        th1l.append(th1)
        y_h = np.zeros(m)
        y_h = th0 + th1 * x
        y_p.append(y_h)
        print("h(x)\n",y_h)
        E = y_h - y
        print("Error Vector :\n",E)
        cost = np.sum(((E)**2))
        s0 = np.sum(E)
        s1 = np.sum(E*x)
        gth0 = (s0/m) 
        gth1 = (s1/m)
       # print("g0\n",gth0)
        #print("g1\n",gth1)
        final_cost = (1/(2*m))*cost
        cost_list.append(final_cost)
        print("j = ",final_cost)
        grad_vec = np.array([gth0,gth1])
        print("Gradiant Vector = \n",grad_vec)
        grad_norm = np.linalg.norm(grad_vec)
        print("Gradiant Norm = \n",grad_norm)
        th0 = th0 - a * gth0
        th1 = th1 - a * gth1
        print("theta0 _ new = ",th0)
        print("theta1 _ new = ",th1)
        if grad_norm < 0.001: 
            
            break
            
    return i,th0,th1,E,final_cost,y_h,cost_list,th0l,th1l,y_p
       
    
```

```python
alpha = 0.0005
iterations = 300
i,th0,th1,E,final_cost,y_h,list_cost,th0l,th1l,y_p = batch(X,y,alpha,iterations)
print("****************** Training Report ********************")
print("Gradient Descent converged after ",i," epochs")
print("theta_0_Opt : ",th0)
print("theta_1_Opt : ",th1)
print("Error Vector :\n",E)
print("cost = ",final_cost)
print("h(x) = y_predict:\n",y_h)
print("y_actual:\n",y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_h)
```

```python
def Adam(x,y,a,e,b1,b2,mx):
    
    #Initialization theta ,v (momentum)
    th0 = 0
    th1 = 0
    mt0 = 0
    mt1 = 0
    vt0 = 0
    vt1 = 0

    alpha = a
    epsillon = e
    beta1 = b1
    beta2 = b2
    m = y.shape[0]
    cost_list = []
    th0l = []
    th1l = []
    y_p = []
    #looping in maxumum number of iterations
    for i in range(mx):
        
        print("****************",i,"**************")
        th0l.append(th0)
        th1l.append(th1)
        # Calculate yhat for theta
        y_h = np.zeros(m)
        y_h = th0 + th1 * x
        y_p.append(y_h)
        print("h(x)\n",y_h)
        
        # Calculate Error Vector for yhat
        E_actual = y_h - y
        print("Error Vector :\n",E_actual)
        #Calculate Cost_function for yhat
        cost = np.sum(((E_actual@E_actual)))
        final_cost = (cost/(2*m))
        cost_list.append(final_cost)
        print("j = ",final_cost)
        
        # Calculate Gradiant for Error Vector for yhat
        s0= np.sum(E_actual)
        s1 = np.sum(E_actual@x)
        gth0 = (s0/m) 
        gth1= (s1/m)
        grad_vec = np.array([gth0,gth1])
        print("Gradiant Vector = \n",grad_vec)
        grad_norm = np.linalg.norm(grad_vec)
        print("Gradiant Norm = \n",grad_norm)
        #update  momentum , v ,theta actual
        mt0 = (mt0 * beta1) + (1 - beta1)*(gth0)
        mt1 = (mt1 * beta2) + (1 - beta2)*(gth1)
        #bais correction for mt
        mth0 = mt0 / (1 - beta1)
        mth1 = mt1 / (1 - beta2)

        vt0 = (beta2 * vt0) + (1 - beta2)*(gth0**2)
        vt1 = (beta2 * vt1) + (1 - beta2)*(gth1**2)
        #bais correction for vt
        vth0 = vt0 / (1 - beta2)
        vth1 = vt1 / (1 - beta2)
        #update theta 
        th0 = th0  - ((alpha /((np.sqrt(vth0)+epsillon)) * mth0))
        th1 = th1  - ((alpha / ((np.sqrt(vth1)+epsillon)) * mth1))
        
        
        print("v0  = ",vth0)
        print("v1 ",vth1)
        
        print("theta0 _ new = ",th0)
        print("theta1 _ new = ",th1)
        #stop cirtira 
        if i > 0 :
            diff = np.abs(cost_list[-1] - cost_list[-2]) 
            if grad_norm < 0.001 or diff < 0.001:
                #print("erroerrrr")
                break
                
        else :
            if grad_norm < 0.0001:
               
                break
            
 
    return i,th0,th1,E_actual,final_cost,y_h,cost_list,th0l,th1l,y_p
       
```

```python
alpha = 0.05
beta1 = 0.05
beta2 = 0.7
epsilon = 1e-08
iterations = 500
i,th0,th1,E,final_cost,y_h,list_cost,th0l,th1l,y_p = Adam(X,y,alpha,epsilon,beta1,beta2,iterations)
print("****************** Training Report ********************")
print("Gradient Descent converged after ",i," epochs")
print("theta_0_Opt : ",th0)
print("theta_1_Opt : ",th1)
print("Error Vector :\n",E)
print("cost = ",final_cost)
print("h(x) = y_predict:\n",y_h)
print("y_actual:\n",y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_h)
```
