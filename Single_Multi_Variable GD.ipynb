### Step6: Repeat from 2 to 5 until converge to the minimum or achieve maximum iterations.
#### The objective from this step is to combine all the previous steps and iterate untill you either achieve the maximum number of iterations or reach the minimum.

```python
def batch(x , y ,alpha,eps= 0.01 ):
    th0 = 0
    th1 = 0
    
    m = X.shape[0]
    y_hat = np.zeros(m)
    #print(y_hat)
    #print(m)
    cost_list = []
    i = 0
    while True:
        
       
        print("****************",i,"**************")

        for j in range(m):
            
            y_hat[j] = th0 + X[j]*th1
        E = y_hat - y
        print("Error Vector = ",E)
        cost_sum = 0
        s0 = 0
        s1 = 0
        for j in range(m):
            
            cost = (y_hat[j] - y[j])**2
            cost_sum += cost
            s0 += E[j]
            s1 += E[j] * X[j]

        gth0 = (s0/m)
        gth1 = (s1/m)

        final_cost = (1/(2*m))*cost_sum
        cost_list.append(final_cost)
        print("j = ",final_cost)
        grad_vec = np.array([gth0,gth1])
        print("Gradiant Vector = \n",grad_vec)
        grad_norm = np.linalg.norm(grad_vec)
        print("Gradiant Norm = \n",grad_norm)

        th0 = th0 - alpha * gth0
        th1 = th1 - alpha * gth1
        print("theta0 _ new = ",th0)
        print("theta1 _ new = ",th1)
        
        if i > 0 :
            diff = np.abs(cost_list[-1] - cost_list[-2])
            # convergence check (cost check)  and stop criteria (Gradient Check)
            if diff < eps or grad_norm < eps:
                
                print("converged after ",i," iterations")
                print("theta0 _ new = ",th0)
                print("theta1 _ new = ",th1)
                break
               
        i = i +1
    return i,th0,th1,E,final_cost,y_hat , cost_list
            





```
