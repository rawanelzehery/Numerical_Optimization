```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
X = np.linspace(0,20)
X
```

```python
a=-1
b=2
y=a*X+b
y
```

```python
plt.scatter(X,y)
```

```python
def batch(x,y,a,mx):
    th0 = 0
    th1 = 0
    m = y.shape[0]
    cost_list = []
    th0l = []
    th1l = []
    y_p = []
    for i in range(mx):
        
        print("****************",i,"**************")
        th0l.append(th0)
        th1l.append(th1)
        y_h = np.zeros(m)
        y_h = th0 + th1 * x
        y_p.append(y_h)
        print("h(x)\n",y_h)
        E = y_h - y
        print("Error Vector :\n",E)
        cost = np.sum(((E)**2))
        s0 = np.sum(E)
        s1 = np.sum(E*x)
        gth0 = (s0/m) 
        gth1 = (s1/m)
       # print("g0\n",gth0)
        #print("g1\n",gth1)
        final_cost = (1/(2*m))*cost
        cost_list.append(final_cost)
        print("j = ",final_cost)
        grad_vec = np.array([gth0,gth1])
        print("Gradiant Vector = \n",grad_vec)
        grad_norm = np.linalg.norm(grad_vec)
        print("Gradiant Norm = \n",grad_norm)
        th0 = th0 - a * gth0
        th1 = th1 - a * gth1
        print("theta0 _ new = ",th0)
        print("theta1 _ new = ",th1)
        if grad_norm < 0.001: 
            
            break
            
    return i,th0,th1,E,final_cost,y_h,cost_list,th0l,th1l,y_p
       
    
```

```python
alpha = 0.0005
iterations = 300
i,th0,th1,E,final_cost,y_h,list_cost,th0l,th1l,y_p = batch(X,y,alpha,iterations)
print("****************** Training Report ********************")
print("Gradient Descent converged after ",i," epochs")
print("theta_0_Opt : ",th0)
print("theta_1_Opt : ",th1)
print("Error Vector :\n",E)
print("cost = ",final_cost)
print("h(x) = y_predict:\n",y_h)
print("y_actual:\n",y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_h)
```

```python
def momuntum(x,y,a,g,mx):
    th0 = 0
    th1 = 0
    vth0 = 0
    vth1 = 0
    gamma = g
    alpha = a
    m = y.shape[0]
    cost_list = []
    th0l = []
    th1l = []
    y_p = []
    for i in range(mx):
        
        print("****************",i,"**************")
        th0l.append(th0)
        th1l.append(th1)
        y_h = np.zeros(m)
        y_h = th0 + th1 * x
        y_p.append(y_h)
        print("h(x)\n",y_h)
        E = y_h - y
        print("Error Vector :\n",E)
        cost = np.sum(((E@E)))
        s0 = np.sum(E)
        s1 = np.sum(E@x)
        gth0 = (s0/m) 
        gth1 = (s1/m)
        
        print("v_0 :",vth0)
        print("v_1 :",vth1)
        print("g0\n",gth0)
        print("g1\n",gth1)
        final_cost = (cost/(2*m))
        cost_list.append(final_cost)
        print("j = ",final_cost)
        grad_vec = np.array([gth0,gth1])
        print("Gradiant Vector = \n",grad_vec)
        grad_norm = np.linalg.norm(grad_vec)
        print("Gradiant Norm = \n",grad_norm)
        vth0 = gamma * (vth0) + alpha * gth0
        vth1 = gamma * (vth1) + alpha *gth1
        th0 = th0 - vth0
        th1 = th1 - vth1
        
        print("theta0 _ new = ",th0)
        print("theta1 _ new = ",th1)
        if i > 0 :
            diff = np.abs(cost_list[-1] - cost_list[-2]) 
            if grad_norm < 0.001 or diff < 0.001:
                print("erroerrrr")
                break
                
        else :
            if grad_norm < 0.0001:
               
                break
            
 
    return i,th0,th1,E,final_cost,y_h,cost_list,th0l,th1l,y_p
       
```

```python
alpha = 0.0005
gamma=0.5 
iterations = 300
i,th0,th1,E,final_cost,y_h,list_cost,th0l,th1l,y_p = momuntum(X,y,alpha,gamma,iterations)
print("****************** Training Report ********************")
print("Gradient Descent converged after ",i," epochs")
print("theta_0_Opt : ",th0)
print("theta_1_Opt : ",th1)
print("Error Vector :\n",E)
print("cost = ",final_cost)
print("h(x) = y_predict:\n",y_h)
print("y_actual:\n",y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,y_h)
```
